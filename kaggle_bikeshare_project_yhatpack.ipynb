#Forecasting Bike Share Demand

"Bike sharing systems are a means of renting bicycles where the process of obtaining membership, rental, and bike return is automated via a network of kiosk locations throughout a city. Using these systems, people are able rent a bike from a one location and return it to a different place on an as-needed basis. Currently, there are over 500 bike-sharing programs around the world.

The data generated by these systems makes them attractive for researchers because the duration of travel, departure location, arrival location, and time elapsed is explicitly recorded. Bike sharing systems therefore function as a sensor network, which can be used for studying mobility in a city. In this competition, participants are asked to combine historical usage patterns with weather data in order to forecast bike rental demand in the Capital Bikeshare program in Washington, D.C." -Kaggle

Data Fields

* datetime - hourly date + timestamp  
* season -  1 = spring, 2 = summer, 3 = fall, 4 = winter 
* holiday - whether the day is considered a holiday
* workingday - whether the day is neither a weekend nor holiday
* weather - 
  * 1: Clear, Few clouds, Partly cloudy, Partly cloudy 
  * 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist 
  * 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds 
  * 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
* temp - temperature in Celsius
* atemp - "feels like" temperature in Celsius
* humidity - relative humidity
* windspeed - wind speed
* casual - number of non-registered user rentals initiated
* registered - number of registered user rentals initiated
* count - number of total rentals

##The Problem
We are given hourly rental data spanning two years. The training set is comprised of the first 19 days of each month, while the test set is the 20th to the end of the month. The problem is to predict the total count of bikes rented during each hour.The prediction is evaluated by reducing the root mean squared log error (RMSLE) of the predicted and actual values. RMSLE is similar to RMSE; however, by using the log error, we reduce the penalty for predicitons of larger values. 

$$RMSLE = \sqrt{\frac{1}{n} \sum_{i=1}^n (\log(p_i + 1) - \log(a_i+1))^2 }$$
Where

* n is the number of hours in the test set
* pi is your predicted count
* ai is the actual count
* log(x) is the natural logarithm$$

# Dev Split Approaches

Machine learning experiments should ideally be run executed using three distinct sets of data:

* Train: The data that is used to fit, or train the model.  This typically includes 80% of the available labeled data (data with outcome values as well as feature values).
* Dev:   The data that is used to initially assess the accuracy of the model.  This typically includes a smaller portion of the 20% remaining of labeled data.
* Test:  The data that is used to perform a final evaluation of the model.  This typically includes that large protion of the remaining 20% of labeled data.

In this competition, the role of the Test data is slightly different as it is not provided with label data to allow us to evaulate accuracy.  The labels are held out by the competition administrators who use it to judge submisssions.  The approaches used to generate a train and dev subset of the provided train data is described below.

##Initial random approach
At first, the train data set was split into train and dev using a standard 80/20 split algorithm.  However, it was quickly realized that this would not work for the time sequential nature of this data.  Additionally, test was already separated from the initial train data in this manner:

* Train: Day 1 - 19, for each month
* Test:  Day 20 - last day of month, for each month

##Static approach
Our next method for splitting the data took an additional few days at the end of each training period for each month and assigned them to the dev set:  

* Train (after split):  Day 1 - 16, for each month  
* Dev (after split):    Day 17 - 19, for each month

##Random but sequential approach
In order to help reduce overfitting our models to the specific train and dev sets described above, we wrote another split algorithm that would include randomness, but still preserved the required sequential aspect of the data.  For each month, a random number of days between a specified range would be assigned to dev from the original train set.  This way, each month would have a variable number of days in train and dev, and each time the data is split, a different train and dev data set would be produced.  Once this algorithm was in place, most model testing was done with the following splits:

* Train (after split):  Day 1 - (20-random number of days between 10 and 2), for each month
* Dev (after split):    Day (20-random number of days between 10 and 2) - 19, for each month

Evaluation using this approach usually involved executing the model for 10 'spliterations' and then averaging the results.


```python
def split_monthly(data, labels, log_labels, day_holdout):
  """
    For each month in original dataset, splits data into test and dev
      test: Data that comes before 20th - day_holdout DAY in month
      dev: Data that comes on or after 20th - day_holdout DAY in month
  """
  
  train_list, train_label_list, train_log_label_list = [],[],[]
  dev_list, dev_label_list, dev_log_label_list = [],[],[]
  for i in range(len(data)):
    if data[i][DAY] < 20-day_holdout:
      train_list.append(data[i])
      train_label_list.append(labels[i])
      train_log_label_list.append(log_labels[i])
    else:
      dev_list.append(data[i])
      dev_label_list.append(labels[i])
      dev_log_label_list.append(log_labels[i])
  
  train_data = np.array(train_list)
  train_labels = np.array(train_label_list)
  train_log_labels = np.array(train_log_label_list)
  
  dev_data = np.array(dev_list)
  dev_labels = np.array(dev_label_list)
  dev_log_labels = np.array(dev_log_label_list)
  
  return train_data, train_labels, train_log_labels, dev_data, dev_labels, dev_log_labels
    
def split_monthly_random(data, labels, log_labels, max_day_holdout, min_day_holdout):
    """
    For each month in original train, splits data into test and dev
    test: data that comes before random of the max holdout day in month
    dev: data that comes on or after day of random max holdout in month
    """
    train_list, train_label_list, train_log_label_list = [],[],[]
    dev_list, dev_label_list, dev_log_label_list = [],[],[]
    
    split_days = [20 - randrange(max_day_holdout) - min_day_holdout for m in range(1,13)]
    
    
    for i in range(len(data)):
        if data[i][DAY] < split_days[int(data[i][MONTH])-1]:
            train_list.append(data[i])
            train_label_list.append(labels[i])
            train_log_label_list.append(log_labels[i])
        else:
            dev_list.append(data[i])
            dev_label_list.append(labels[i])
            dev_log_label_list.append(log_labels[i])
    
    train_data = np.array(train_list)
    
    train_labels = np.array(train_label_list)
    train_log_labels = np.array(train_log_label_list)
  
    dev_data = np.array(dev_list)
    
    dev_labels = np.array(dev_label_list)
    dev_log_labels = np.array(dev_log_label_list)
  
    return train_data, train_labels, train_log_labels, dev_data, dev_labels, dev_log_labels
```

##Linear Regression
We started modeling with a simple __Ordinary Least Squares__ linear regression model. 

Linear regression fits a linear model to minimize the residual sum of squares between the observed responses in the dataset, and the responses predicted by the linear approximation. However, coefficient estimates for Ordinary Least Squares rely on the independence of the model terms. When terms are correlated, the least-squares estimate becomes highly sensitive to random errors in the observed response, producing a large variance.

__Ridge regression__ addresses some of the problems of Ordinary Least Squares by imposing a penalty on the size of coefficients. 

The __Lasso__ is a linear model that estimates sparse coefficients. It is useful in some contexts due to its tendency to prefer solutions with fewer parameter values, effectively reducing the number of variables upon which the given solution is dependent.

##Random Forest
__Decision trees__ identify features that are most discerning when it comes to identifying classes. The construction of the tree is built using a recursive partition (see CART), which essesntially means .... The method searches for all distinct predictors, and splits into groups that minimize a loss function (RMSLE in our case). The top of the tree, root node, represents the best predictor. In contrast to linear regression models that calculate coefficients of predictors, tree regression models calculate the relative importance of predictors. The importance of each predictor is computed by summing the reduction the overall reduction of the omptimization criteria like RMSLE.

A __random forest__ is a collection of these decision trees. The forest approach relies on an average of many small trees without analyst interacti. The idea is that the errors from these small trees are overcome with the aggreagted best predictor. Benefits of a random tree include insensitivity to unimportant features.


###Bagging and Boosting
__Bagging__ involves having each model in the ensemble vote with equal weight. In order to promote model variance, bagging trains each model in the ensemble using a randomly drawn subset of the training set.

__Boosting__ involves incrementally building an ensemble by training each new model instance to emphasize the training instances that previous models mis-classified. In some cases, boosting has been shown to yield better accuracy than bagging, but it also tends to be more likely to over-fit the training data. By far, the most common implementation of Boosting is Adaboost, although some newer algorithms are reported to achieve better results.

## AdaBoost Regression

The core principle of AdaBoost is to fit a sequence of weak learners (i.e., models that are only slightly better than random guessing, such as small decision trees) on repeatedly modified versions of the data. The predictions from all of them are then combined through a weighted majority vote (or sum) to produce the final prediction.

We did not experiment too much with AdaBoost, as a few attempts at tuning the hyperparameters did not bring us close to where we were with random forest regression.

Hyperparameters are a set of controls an analyst can use to optimize a learning algorithm. Obraining the specific value is more of an art than science. Pruning a decision tree is an example of a hyperparameter that prevents overfitting. Other parameters include the depth of the tree, which is used as a penalty 

__Hyperparameter Tuning for Gradient Boosting:__

The size of the regression tree base learners defines the level of variable interactions that can be captured by the gradient boosting model. In general, a tree of depth h can capture interactions of order h . If you specify max_depth=h then complete binary trees of depth h will be grown. Such trees will have (at most) 2**h leaf nodes and 2**h - 1 split nodes.

Alternatively, you can control the tree size by specifying the number of leaf nodes via the parameter max_leaf_nodes. In this case, trees will be grown using best-first search where nodes with the highest improvement in impurity will be expanded first. A tree with max_leaf_nodes=k has k - 1 split nodes and thus can model interactions of up to order max_leaf_nodes - 1 .

We found that max_leaf_nodes=k gives comparable results to max_depth=k-1 but is significantly faster to train at the expense of a slightly higher training error.

learning_rate is a hyper-parameter in the range (0.0, 1.0]. learning_rate strongly interacts with the parameter n_estimators, the number of weak learners to fit. Smaller values of learning_rate require larger numbers of weak learners to maintain a constant training error. Empirical evidence suggests that small values of learning_rate favor better test error. recommend to set the learning rate to a small constant (e.g. learning_rate <= 0.1) and choose n_estimators by early stopping.

# Feature Engineering

 human-perceived equivalent temperature##Throughout this exercise we have experimented with adding (and subtracting) a number of features to our model in order to improve our prediction score. Below is a representative list of those features and the motivation behind them:
###Additions:
  * __Weekday:__ Certain days of the week would have higher rentals than others (For example, casual is high for Saturdays and Sundays).
  * __3-day weekends:__ People go on vacation on 3-day weekends, and so there could be more demand for bikes on those days.
  * __Wet grounds:__ People avoid biking on days when there are puddles of water on the ground.
  * __Past temperature:__ Temperature of the past hour can have an effect on the decision to bike for casual bikers.
  * __Past atemp:__ Same as above.
  * __High windspeed:__ Windspeed may not affect bike demand unless it reaches a certain threshold.  Experimentation indicated 35mph was an effective threshold.
  * __Average hour/weekday:__ There may be a pattern that occurs where a similar bike demand exists for the same hour on the same weekday.  This feature calculates a 3 month moving average of all the previous counts available for the same weekday and hour, for each row of data.
  * __MA:__ High demand for bikes during one hour would create a shortage for bikes in the following hours. More on this in the ARIMA section below.
  * __Heat Index:__ Human perceived temperature that takes humidity and temperature into account

###Subtractions:
  * __Month:__ because milad said so, but i may have mishead, anywho, it boosted our place on the leaderboard, and that's all that matters
  * __Day of month:__ Bike rentals should not be correlated with day of month. For example, the fact that today is the 14th of the month wouldn't have an effect on bike rentals.

###Discarded:
  * __Recoding categorical features:__ One idea was that a decision tree might have an easier time making splitting decisions if the categories in a feature are ordered according to the output.
  * __Humidity < 16:__ Humidity distribution looks funny for values under 16.
  * __Combine holiday and workingday:__ We tried running a Random Forest model with our 83-feature dummy-variabled data set and got really bad results. So naturally we thought maybe combining feature might help Random Forest.
  * __Transforming and mean-normalizing continuous variables:__ transforming and normalizing temp, atemp, windspeed, or humidity did not improve our models.
  * __Subtracting windspeed:__ This was removed when the high-windspeed feature was added, but it resulted in worse results.


#Autoregressive Integrated Moving Average - ARIMA
We thought our model would benefit from using ARIMA techniques that would allow a smoothed, moving average of known outcomes to be used as features for predicting test outcomes.  This was highly recommended to look into considering our challenge is largely a time series analysis.
##Manual calculation of moving averages
The statspackage for Python implements functionality for ARIMA.  This was investigated a little, but there was little documentation.  It was decided to manually build our own functions for calculating moving average.  The moving average for each of the output variables (casual count, registered count, and overall count) was calculated and added as features to the model.  This worked well when evaluating test models, as the labels to calculate the features are available in the test and dev sets.  However, it is problematic when running the model against test, since all features in the train data used to fit the model must exist in the test data that is used for prediction.  There were two approaches that were explored for dealing with this issue.

###One-at-a-time prediction
Given that each test window follows a train/dev window, we can calculate the average output of the past 2 hours for each observation in test one by one. We start with the first observation in test, and look at the average of the last 2 observations in train/dev. The model would then predict the outcome for that observation. For the next observation, we can then use our prediction from the last observation. This process continues until we have predictions for all test observations.

###Double prediction
The problem could be addessed in a manner similar to the Expectation Maximization algorithm, where initial values are chosen for the outcome, and then the model iterative works from that starting point to achieve best results.  Here, the model was built with all the normal features (but not the moving averages), fit against train, and then used to predict the outcome variables for test.  The model same model would then be built again, but this time include the moving average features.  These features could now be added to the test data, as the outcomes predicted in the first step could be used.  The model could now output a final set of predictions that incorporated the moving average features.

The first attempt using this approach led to the team's current best score.

###Multiple prediction
Several attempts were made where the double prediciton approach was used, but instead of just one additional pass, the model was executed multiple time, each time feeding in the previous predictions in the moving average calcualtions, hoping to converge on more accurate results.  The evidence suggested this approach did not improve the outcomes.
